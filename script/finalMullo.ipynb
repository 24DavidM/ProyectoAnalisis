```python
#PASAMOS EL PRIMER DATAFRAME A BASE RELACIONAL MYSQL WORKBENCH
import sqlalchemy
from sqlalchemy import create_engine
usuario = 'root'
contrasenia = '123456'
host = 'localhost'
puerto = '3306'
base_de_datos = 'base_no_relacional'
```

```python
# Conexion a MYSQL
connection_string = f'mysql+mysqlconnector://{usuario}:{contrasenia}@{host}:{puerto}/{base_de_datos}'
engine = create_engine(connection_string)
```

```python
batch_size = 5000

try:

    with engine.begin() as connection:
   
        for start in range(0, len(data1), batch_size):
            batch = data1.iloc[start:start + batch_size]
            try:
                # Insertar el lote en la base de datos
                batch.to_sql('dosk_registros', con=connection, index=False, if_exists='append') 
                print(f"Lote {start // batch_size + 1} insertado exitosamente.")
            except Exception as e:
                print(f"Error al insertar el lote {start // batch_size + 1}: {e}")
                connection.rollback()
                break
except Exception as e:
    print(f"Error de conexiÃ³n o transacciÃ³n: {e}")
finally:
    engine.dispose()
```

```python
#Se establece la conexion pero para la nueva base de datos
usuario = 'root'
contrasenia = '123456'
host = 'localhost'
puerto = '3306'
base_de_datos = 'champions'

# Crear la cadena de conexiÃ³n y el engine
connection_string = f'mysql+mysqlconnector://{usuario}:{contrasenia}@{host}:{puerto}/{base_de_datos}?connect_timeout=300'
engine = create_engine(connection_string)

data3 = data3.copy()
data3.rename(columns={"weight(kg)": "weight_kg", "height(cm)": "height_cm"}, inplace=True)
```

```python
#Ahora insertaremos los datos relacionados a mysql
with engine.begin() as connection:
    try:
        # Insertar en la tabla equipos
        data2.to_sql('equipos', con=connection, index=False, if_exists='append')
        # Luego insertar en jugadores
        data3.to_sql('jugadores', con=connection, index=False, if_exists='append')
        # Finalmente insertar en goles
        data4.to_sql('goles', con=connection, index=False, if_exists='append')

        print("DataFrames insertados exitosamente.")
    except Exception as e:
        print(f"Error durante la operaciÃ³n de la base de datos: {e}")
```
