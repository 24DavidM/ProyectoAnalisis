```python
# TECER DATAFRAME DE LOS EVENTOS ATLETICOS
```

```python
dfAthlete = pd.read_csv("./csv/athlete_events.csv", index_col="ID")
```

```python
dfAthlete.head()
```

```python
# Verificar si hay null
dfAthlete.isnull()
```

```python
# Contar valores nulos por cada columna
dfAthlete.isnull().sum()
```

```python
# Remplaza en Medalla con poner No medalla lo que no ganaron
dfAthlete['Medal'] = dfAthlete['Medal'].fillna('Not medal')
```

```python
# Remplaza la edad para poner el promedio de edad en general
dfAthlete['Age'] = dfAthlete['Age'].fillna(dfAthlete['Age'].mean())
```

```python
# Remplaza la edad de tipo decimal a entero
dfAthlete['Age'] = dfAthlete['Age'].astype(int)
```

```python
# Remplaza el peso para poner el promedio del peso en general
dfAthlete['Weight'] = dfAthlete['Weight'].fillna(dfAthlete['Weight'].mean())
```

```python
dfAthlete['Weight'] = dfAthlete['Weight'].round(2)
```

```python
# Remplaza la altura para poner el promedio de altura en general
dfAthlete['Height'] = dfAthlete['Height'].fillna(dfAthlete['Height'].mean())
```

```python
dfAthlete['Height'] = dfAthlete['Height'].round(2)
```

```python
#Se reduce valores nulos
dfAthlete.isnull().sum()
```

```python
# Guardamos dataframe en un csv
dfAthlete.to_csv('./limpiosCsv/CleanAthlete.csv', encoding='utf-8')
```
