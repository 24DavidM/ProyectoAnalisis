```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cancer_de_mama_limpio.csv', header=None, names=["id", "diagnosis", "radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean", "compactness_mean", "concavity_mean", "concave_points_mean", "symmetry_mean", "fractal_dimension_mean", "radius_se", "texture_se", "perimeter_se", "area_se", "smoothness_se", "compactness_se", "concavity_se", "concave_points_se", "symmetry_se", "fractal_dimension_se", "radius_worst", "texture_worst", "perimeter_worst", "area_worst", "smoothness_worst", "compactness_worst", "concavity_worst", "concave_points_worst", "symmetry_worst", "fractal_dimension_worst"])
df['diagnosis'] = df['diagnosis'].map({'M': 'Maligno', 'B': 'Benigno'})
plt.figure(figsize=(6, 4))
sns.countplot(x='diagnosis', data=df, palette='Set2')
plt.title('Distribución de Tumores Benignos y Malignos')
plt.xlabel('Diagnóstico')
plt.ylabel('Número de Casos')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('cancer_de_mama_limpio.csv')

# Calcular el Radio Promedio para los casos Malignos (M) y Benignos (B)
radio_promedio_maligno = df[df['diagnosis'] == 'M']['radius_mean'].mean()
radio_promedio_benigno = df[df['diagnosis'] == 'B']['radius_mean'].mean()

# Etiquetas con los valores de radio promedio
labels = [f'Maligno: {radio_promedio_maligno:.2f}', f'Benigno: {radio_promedio_benigno:.2f}']

# Valores para el gráfico
sizes = [radio_promedio_maligno, radio_promedio_benigno]

# Crear un gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Distribución del Radio Promedio de Cáncer Maligno (M) y Benigno (B)', fontsize=14)
plt.axis('equal')  # Para que el gráfico sea un círculo perfecto
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('cancer_de_mama_limpio.csv')

# Calcular el área promedio de los tumores malignos (M) y benignos (B)
area_peor_maligno = df[df['diagnosis'] == 'M']['area_worst'].mean()
area_peor_benigno = df[df['diagnosis'] == 'B']['area_worst'].mean()

# Etiquetas con los valores del área peor
labels = [f'Maligno: {area_peor_maligno:.2f}', f'Benigno: {area_peor_benigno:.2f}']

# Valores para el gráfico
sizes = [area_peor_maligno, area_peor_benigno]

# Crear un gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Distribución del Área Peor de Cáncer Maligno (M) y Benigno (B)', fontsize=14)
plt.axis('equal') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('cancer_de_mama_limpio.csv')

# Calcular la suavidad peor para los tumores malignos y benignos
suavidad_peor_maligno = df[df['diagnosis'] == 'M']['smoothness_worst'].mean()
suavidad_peor_benigno = df[df['diagnosis'] == 'B']['smoothness_worst'].mean()

# Datos para el gráfico
labels = ['Maligno', 'Benigno']
suavidad_peor = [suavidad_peor_maligno, suavidad_peor_benigno]

# Crear el gráfico de barras para la suavidad peor
plt.figure(figsize=(8, 6))
plt.bar(labels, suavidad_peor, color=['lightcoral', 'lightgreen'])
plt.title('Suavidad Peor', fontsize=14)
plt.ylabel('Valor de Suavidad Peor', fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('cancer_de_pulmon_limpio.csv')

# Contar las ocupaciones más comunes
ocupaciones_comunes = df['Occupation'].value_counts().head(5)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
ocupaciones_comunes.plot(kind='bar', color='skyblue')

# Título y etiquetas
plt.title('Las 5 Ocupaciones Más Comunes', fontsize=16)
plt.xlabel('Ocupación', fontsize=12)
plt.ylabel('Cantidad de Casos', fontsize=12)

# Mostrar el gráfico
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()

# Imprimir los resultados
print(ocupaciones_comunes)



```
